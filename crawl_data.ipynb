```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import requests
import pandas as pd
import math
import re
print("Importing done")
```

```python
driver = webdriver.Chrome()
url = 'https://www.sciencedirect.com/browse/journals-and-books?contentType=JL'
driver.get(url)
sleep(3)
```

```python
count_journals  = driver.find_element('xpath', '//*[@id="main-content"]').text
# print('Number of journals: ', count_journals)
for s in count_journals.split():
    try:
        s = (s.replace(',', ''))
    except: pass
    if s.isdigit():
        print('Number of journals: ',s)
        number_of_page = math.ceil(int(s)/100)

print('Number page on website: ',number_of_page)
```

```python

# number_of_page = driver.find_element('xpath', '//*[@id="react-root"]/div/div/div/main/div[3]/div[2]/div/div[1]/div/div/div/span').text
# print('Number page on website: ',number_of_page)

cards = []
arr_link_page = []

for i in range(1, number_of_page+1):
    link_all_page= f'https://www.sciencedirect.com/browse/journals-and-books?page={i}&contentType=JL'
    arr_link_page.append(link_all_page)
    
for link_page in arr_link_page:
    driver.get(link_page)
    requests.get(link_page)
    soup =  BeautifulSoup(driver.page_source)
    tag_cards = soup.find_all('a', 'anchor js-publication-title anchor-default')
    sleep(2)
    cards.append(tag_cards)
```

```python
link_journals= []
for link in cards:
    for tag_a in link:
        link_journals.append('https://www.sciencedirect.com/' + tag_a.get('href'))
```

```python
Sothutu = []
tentapchi = []
issn = []
eissn = []
print_issn = []
online_issn = []
impact_factor =[]
tag_div = []
link_check = []
link_not_found = []
```

```python
for link in link_journals:
    driver.get(link)
    sleep(0.5)
    link_check.append(link)
    try:
        journal_title = driver.find_element('xpath','//*[@id="journal-title"]/span').text
        tentapchi.append(journal_title)
    except:
        tentapchi.append('')
        print('Error title: ', link)
        link_not_found.append(link)

    try: 
        impact = driver.find_element('xpath','//*[@id="react-root"]/div/div/div/main/div[1]/div/section/div[1]/div/div[2]/div[2]/div[2]/button/span/span[1]').text
        # print(impact)
        impact_factor.append(impact)
    except: impact_factor.append('')

    try:
        requests.get(link)
        soup =  BeautifulSoup(driver.page_source)
        div = soup.find_all('div', 'js-issn text-s')
        tag_div.append(div)
    except: 
        print('Error: ', link)
        pass
```

```python

text_list = []
# text_2_list = []
sos = []
# Lặp qua các danh sách con và trích xuất ISSN nếu tồn tại
for div_list in tag_div:
    if not div_list:  # Kiểm tra danh sách rỗng
        text_list.append('')  
    else:
        for div in div_list:
            if div == '':
                text_list.append('')
            div = div.prettify()

            soup = BeautifulSoup(div, 'html.parser')

            div_element = soup.find('div')  # Lấy thẻ div đầu tiên

            vers = div_element.find_all('div')


            contents = [ver.text for ver in vers]   
            if (len(contents))>2:
                print('More than 2 tag: ',contents)
                sos.append(contents)

            text_list.append(contents)
```

```python
import re 

list_moi = []
list_moi_2 = []
list_moi_3 = []
for danh_sach_con in text_list:
    chuoi_con = ''.join(map(str, danh_sach_con))
    chuoi_con = chuoi_con.replace('\n', '')
    chuoi_con = chuoi_con.strip()
    # print((chuoi_con))
    pattern = r'\s{2,}'
    if re.search(pattern, chuoi_con):
        parts = re.split(pattern, chuoi_con)
    # Kiểm tra xem parts có đủ hai phần không trước khi truy cập
        # if len(parts) <= 2:
        chuoi_con_1 = parts[0].strip()
        chuoi_con_2 = parts[1].strip()
        list_moi.append(chuoi_con_1)
        list_moi_2.append(chuoi_con_2)
        if len(parts) == 3:
            chuoi_con_3 = parts[2].strip()
            list_moi_3.append(chuoi_con_3)
        else: list_moi_3.append('')
    else:
        list_moi.append(chuoi_con)
        list_moi_2.append('')
        list_moi_3.append('')
```

```python
# Tạo các cột mới cho các loại tiền tố khác nhau
df_journal['ISSN_1'] = df_journal['issn'].apply(lambda x: x if x.startswith('ISSN') else None) 
df_journal['ISSN_2']=  df_journal['issn_2'].apply(lambda x: x if x.startswith('ISSN') else None)
df_journal['ISSN_3']=  df_journal['issn_3'].apply(lambda x: x if x.startswith('ISSN') else None)
df_journal['Online_1'] = df_journal['issn'].apply(lambda x: x if x.startswith('Online') else None) 
df_journal['Online_2'] = df_journal['issn_2'].apply(lambda x: x if x.startswith('Online') else None)
df_journal['Online_3'] = df_journal['issn_3'].apply(lambda x: x if x.startswith('Online') else None)
df_journal['Print_1'] = df_journal['issn'].apply(lambda x: x if x.startswith('Print') else None) 
df_journal['Print_2'] = df_journal['issn_2'].apply(lambda x: x if x.startswith('Print') else None)
df_journal['Print_3'] = df_journal['issn_3'].apply(lambda x: x if x.startswith('Print') else None)
df_journal['e_1'] = df_journal['issn'].apply(lambda x: x if x.startswith('e') else None) 
df_journal['e_2'] =  df_journal['issn_2'].apply(lambda x: x if x.startswith('e') else None)
df_journal['e_3'] =  df_journal['issn_3'].apply(lambda x: x if x.startswith('e') else None)
df_journal['Linking_1'] = df_journal['issn'].apply(lambda x: x if x.startswith('Linking') else None) 
df_journal['Linking_2'] =  df_journal['issn_2'].apply(lambda x: x if x.startswith('Linking') else None)
df_journal['Linking_3'] =  df_journal['issn_3'].apply(lambda x: x if x.startswith('Linking') else None)



df_journal
```
